```python
def Function(x):
    return 5*(1-np.exp(-x))-x
```

```python
def Derivative(f,x,h=1e-6):
    return (f(x+h)-f(x-h))/(2*h)
```

```python
def GetNewtonMethod(f,df,xn,itmax=1000,precision=1e-5):
    
    error = 1
    it=0
    
    while error > precision and it < itmax:
        
        try:
            
            xn1 = xn - f(xn)/df(f,xn)
            error = np.abs(f(xn)/df(f,xn))
            
        except ZeroDivisionError:
            print('Division por cero')
            
        xn = xn1
        it += 1
        
    print('raiz:', xn,it)
    
    if it == itmax:
        return False
    else:
        return xn
```

```python
root = GetNewtonMethod(Function,Derivative,1)
```

```python
def GetAllRoots(x,tolerancia=8):
    
    Roots = np.array([])
    
    for i in x:
        root = GetNewtonMethod(Function,Derivative,i)
        
        if root != False:
            
            croot = np.round( root, tolerancia )
            
            if croot not in Roots:
                Roots = np.append(Roots,croot)
                
    Roots.sort()
    
    return Roots
```

```python
xtrial = np.linspace(-10,10,100)
Roots = GetAllRoots(xtrial)
```
